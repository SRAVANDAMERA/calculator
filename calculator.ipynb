```python
a=int(input("enter the first number : "))
b=int(input("enter the second number : "))
result=0
operat=input("enter any arithematic operation : ")

if operat =='+':
  result = a + b
  print("addition of a & b is : ",result)

elif operat =='-':
  result = a - b
  print("subtraction of a & b is : ",result)

elif operat =='*':
  result = a * b
  print("multiplication of a & b is : ",result)

elif operat =='/':
  result = a / b
  print("division of a & b is : ",result)

elif operat =='%':
  result = a % b
  print("modulus of a & b is : ",result)

else:
  print("enter a valid operator")


```
